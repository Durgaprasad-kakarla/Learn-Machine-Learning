```python
from io import StringIO,BytesIO
```

```python
data=('col1,col2,col3\n'
     'x,y,1\n'
     'a,b,2\n'
     'c,d,3\n')
```

```python
type(data)
```

```python
pd.read_csv(StringIO(data))#stringIO() method is used to retrieve data in that "data" string
```

```python
df=pd.read_csv(StringIO(data),dtype=object)#it will make the data as objects
df
```

```python
data1=('a,b,c,d\n'
      '1,2,3,4\n'
      '5,6,7,8\n'
      '9,10,11')
```

```python
pd.read_csv(StringIO(data1))
```

```python
df=pd.read_csv(StringIO(data1),dtype=object)
```

```python
df
```

```python
df1=pd.read_csv(StringIO(data1),dtype=float)
df1
```

```python
df1.isnull().sum()
```

```python
df=pd.read_csv(StringIO(data1),dtype={'b':int,'c':float,'a':'Int64'})
df
```

```python
type(df)
```

```python
pd.read_csv(StringIO(data1),index_col=1)
```
