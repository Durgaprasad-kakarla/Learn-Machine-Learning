### one hot encoding with many categories in a feature 

```python
df=pd.read_csv('mercedesbenz.csv',usecols=['X0','X1','X2','X3','X4','X5','X6'])
```

```python
df
```

```python
df['X0'].value_counts()
```

```python
for i in df.columns:
    print(len(df[i].unique()))#unique values in each columsn in the df.
```

```python
df['X1'].value_counts().head(10)
```

```python
lst_10=df['X1'].value_counts().head(10).index
lst_10=list(lst_10)
lst_10
```

```python
import numpy as np
for categories in lst_10:
    df[categories]=np.where(df['X1']==categories,1,0)
```

```python
df
```

```python
lst_10.append('X1')
```

```python
df[lst_10].head(50)#instead of creating features from unique values it is better to
#take only top ten features
```
