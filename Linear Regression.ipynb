# Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
#-1 in reshape function is used when you dont know or want to explicitly tell the dimension of that axis. E.g,
#If you have an array of shape (2,4) then reshaping it with (-1, 1),
#then the array will get reshaped in such a way that the resulting array has only 1 column and this is only possible by having 8 rows, hence, (8,1).
x1=df_train.x.values.reshape(-1,1)
y1=df_train.y.values.reshape(-1,1)
print(x1,y1)
```

```python
df_train.dropna(axis=0,inplace=True)
```

```python
df_train.isnull().sum()
```

```python

print(x1.shape)
print(y1.shape)
```

```python
lr.fit(x1,y1)
```

```python
y3=lr.predict(x1)
y3
```

```python
lr.score(x1,y1)
```
