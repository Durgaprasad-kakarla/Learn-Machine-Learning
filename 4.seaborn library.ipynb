# Seaborn:
 ### Seaborn is a library that uses matplotlib underneath to plot graphs.It is used for exploratory data.It will be used to visualize random distributions

### Univariate analysis:
     Analysing one feature in the dataset
### Bivariate analysis:
     Analysing two features in the dataset
### Multivariate analysis:
     Analysing more than two features in the dataset

##### Seaborn can be used to analyse any number features.Seaborn is very useful for exploratory data analysis

# Distribution Plots:
###### 1.Joint plot
###### 2.Pair plot
###### 3.Dist plot

### Pair Plot:Used for multivariate analaysis
### Joint Plot:Used for Bivariate analysis
### Dist Plot:Used for Univariate analysis

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
df.head(100)
```

# correlation:
  Correlation is used to find the interrelationship of features

```python
df.corr()
```

# correlation with heatmap:
A correlation heatmap used colored cells,typically in a monochromatic scale to show a 2D,correlation table between two discrete dimensions.
It is very useful for feature selection

```python
sns.heatmap(df.corr())#here we can observe that total bill and tips are having strong relationship
#tips and size are having weak relationship
```

```python
df.dtypes#dtypes gives the data types of coloumns or features
```

# Joint Plot:
 It is used for bivariate analysis.It is used to study the relationship between two numeric variables.The central chart display their correlation.It is usually a scatterplot,a 2D histogram

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# Pair Plot:
It is used for multivariate analysis
A pair plot is also known as Scatter plot,in which one variable in the same data row is matched with another variable's value,like this pairplots are just eloborations on this,showing all variables paired with all the other variables

```python
sns.pairplot(df)#here more than one feature is compared with each other,so we can find the relation between them.
```

```python
sns.pairplot(df,hue='sex')
```

# Dist Plot:
It is used for univariate analysis
Dist plot helps to check the distribution of the coloumns feature.It is used to create histograms

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)#kde is kernal density estimation is used to estimate the density
```

```python
sns.distplot(df['tip'],kde=False)
```

# Catergorical plots:
Seaborn also helps us in doing the analysis on categorical points.There are different plots in this.They are
### 1.Count Plot
### 2.Bar Plot
### 3.Box Plot
### 4.Violin Plot

# Count Plot:
It is used to count the categorical features.It shows the count of observations in each categorical bin using bars

```python
sns.countplot(x='sex',data=df)#It give the count of male and female
```

```python
sns.countplot(x='smoker',data=df)
```

```python
sns.countplot(y='day',data=df)
```

```python
sns.countplot(x='time',data=df)
```

# Bar Plot
A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

# Box Plot:
A box plot is a graph that presents information from a five number summary

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',data=df)
```

```python
sns.boxplot(x='total_bill',y='day',data=df,hue='smoker')
```

# Violin Plot:
Violin plot helps us to see both the distribution of data in terms of kernal density estimation and the box plot

```python
sns.violinplot(x='total_bill',y='day',data=df)
```

```python
sns.violinplot(x='total_bill',y='day',data=df,hue='smoker')
```
