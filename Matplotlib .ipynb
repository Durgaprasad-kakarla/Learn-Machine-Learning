# Matplotlib:
Matplotlib is a plotting library for the python programming language and its numerical mathematics extension Numpy.It provides object oriented API for embedding plots into applications using general purpose GUI toolkits like wxpython,QT or GTK
Uses:
-generally easy to get started for simple plots
-support for custom labels and texts
-Great control of every element in a figure
-High quality output in many formats
-very customizable in general

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.arange(0,10)
y=np.arange(10,20)
```

```python
print(x)
print(y)
```

```python
plt.scatter(x,y,c='y')
plt.xlabel('X-axis')#It is used to give the label to x-axis
plt.ylabel('Y-axis')#It is used to give the label to y-axis
plt.title('Matplotlib Graph')
```

```python
a=np.arange(5,30)
b=a**2
b
```

```python
plt.scatter(a,b,c='g')#here a contains 25 elements so graph will have 25 points each elements square will be plotted according to x axis
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Square of a number y=x**2')
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Graph')
```

```python
b=a**4
plt.plot(a,b,'r*-')
```

```python
z=3*x+20
plt.plot(x,z,'r*--',linewidth=0.1,markersize=10)
#plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(a,b,'r*-')
plt.subplot(2,2,2)
plt.plot(a,b,'b*-')
plt.subplot(2,2,3)
plt.plot(a,b,'go-')
plt.subplot(2,2,4)
plt.plot(a,b,'yo-')
```

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
x
```

```python
y=np.sin(x)
plt.plot(x,y)
plt.title('Sine wave form')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
#plt.subplot(2,2,1)
plt.plot(x,y_sin)
plt.plot(x,y_cos)
```

```python
plt.subplot(2,2,1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.subplot(2,2,2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x=[2,8,10]
y=[11,16,9]
x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Graph')
plt.show()
```

# Histogram

```python
a=np.array([22,87,5,43,56,73,54,51,5,79,31,65,80,55,66])
plt.hist(a)
plt.title('Histogram')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=False)
```

# Piechart

```python
labels='python','c++','Ruby','Java','c'
sizes=[215,130,245,210,500]
colors=['gold','yellowgreen','lightgrey','blue','lightskyblue']
explode=(0.5,0,0,0,0)
plt.pie(sizes,explode=explode,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
