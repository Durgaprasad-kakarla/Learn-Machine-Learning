### Count or Frequency Encoding

```python
import pandas as pd
df=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header=None)
```

```python
df.head()
```

```python
df[1].unique()
```

```python
columns=[1,3,5,6,7,8,9,13]
```

```python
df=df[columns]
```

```python
df.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
df
```

```python
for col in df.columns[:]:
    print(col,':',len(df[col].unique()),'labels')
```

```python
country_map=df['Country'].value_counts().to_dict()
```

```python
df['Country']=df['Country'].map(country_map)
```

```python
df.head(20)
```

#### Disadvantages

#### Target Guided Ordinal Encoding
1.Ordering the labels according to the target
2.Replace the labels by the joint probability of being 1 or 0

```python
df=pd.read_csv('titanic_train.csv',usecols=['Cabin','Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df.head(20)
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values()
```

```python
ordinal_labels=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_labels2#given the rank to the values
```

```python
df['Cabin_ordinal_labels']=df['Cabin'].map(ordinal_labels2)
df.head()
```

# Mean Encoding

```python
mean_ordinal=df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
df['mean_ordinal_encode']=df['Cabin'].map(mean_ordinal)
```
