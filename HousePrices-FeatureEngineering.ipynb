# Missing values

```python
#First we will handle categorical features
features_with_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']
```

```python
for feature in features_with_nan:
    print(feature,np.round(dataset[feature].isnull().mean(),4))
```

```python
# Replace Missing value with new label
def replace_category_feature(dataset,features_with_nan):
    data=dataset.copy()
    data[features_with_nan]=data[features_with_nan].fillna("Missing")
    return data
dataset=replace_category_feature(dataset,features_with_nan)
dataset[features_with_nan].isnull().sum()
```

```python
dataset.head()
```

#### Numerical features with NaN

```python
numerical_features_with_nan=[feature for feature in dataset.columns if dataset[feature].dtypes!="O" and dataset[feature].isnull().sum()>1]
```

```python
for feature in numerical_features_with_nan:
    print(feature,np.round(dataset[feature].isnull().mean(),4),"% missing values")
```

```python
for feature in numerical_features_with_nan:
    #when we have done exploratory data analysis we found that numerical features are having outliers so we will replace 
    #NaN values with median
    median_value=dataset[feature].median()
    # create  a new feature to capture nan values
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
dataset[numerical_features_with_nan].isnull().sum()
    
```

```python
dataset.head(30)
```

```python
#Temporal variables(Date Time variables)
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature]=dataset['YrSold']-dataset[feature]
```

```python
dataset.head()
```

```python
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']].head(30)
```

```python
num_features=['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']#skewed features
for feature in num_features:
    dataset[feature]=np.log(dataset[feature])
```

# Handling Rare Categorical Feature
we will remove categorical variables that are present less than 1% of the observations

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
```

```python
categorical_features
```

```python
for feature in categorical_features:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```
