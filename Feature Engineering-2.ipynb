## Arbitary value Imputation

 It consists of replaceing NaN by an arbitary value.
 Arbitary is random choice that may good results to us.

```python
import pandas as pd
```

```python
df=pd.read_csv('titanic_train.csv',usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

```python
def impute_nan(df,variable):
    df[variable+"_zero"]=df[variable].fillna(0)
    df[variable+'_hundred']=df[variable].fillna(100)
```

```python
impute_nan(df,'Age')
```

```python
df.head(50)
```

```python
df['Age'].hist(bins=50)
```

## Advantages
- Easy to implement.
- Captures the importance of missingness if there is one
## Disadvantages
- Distorts the original distribution of the variable
- If missingness is not important, it may mask the predictive power of the original variable by distorting its distribution
- Hard to  decide which value to use

```python
def impute_nan(df,variable):
    #Here we replaced the null values with the value which is more frequent 
    most_frequent_category=df[variable].value_counts().index[0]
    df[variable].fillna(most_frequent_category,inplace=True)
```

```python
for feature in df.columns:
    impute_nan(df,feature)
```

```python
df.isnull().sum()
```

```python
df.tail(50)
```

## Advantages
- Easy to implement
- Faster way to implement
## Disadvantages
- Since we are using the more frequent labels, it may use them in an over represented way, if there are many nan's
- It distorts the relation of the most frequent label

### Adding a variable to capture NaN

```python
import numpy as np
```

```python
df['BsmtQual_var']=np.where(df['BsmtQual'].isnull(),1,0)
```

```python
df.head()
```

```python
frequent=df['BsmtQual'].mode()[0]#it is used find the most frequent value
```

```python
df['BsmtQual'].fillna(frequent,inplace=True)
```

```python
# Most probably we have to put the mode if feature has more no of null values
df['FireplaceQu_var']=np.where(df['FireplaceQu'].isnull(),1,0)
frequent=df['FireplaceQu'].mode()[0]
df['FireplaceQu'].fillna(frequent,inplace=True)
```

### suppose if you have more frequent categories, we just replace NaN with a new category

```python
df=pd.read_csv('housepricetrain.csv',usecols=['BsmtQual','FireplaceQu','GarageType','SalePrice'])
```

```python
df.head()
```

```python
def impute_nan(df,variable):
    df[variable+'_newvar']=np.where(df[variable].isnull(),'missing',df[variable])
```

```python
for feature in ['BsmtQual','FireplaceQu','GarageType']:
    impute_nan(df,feature)
```

```python
df.head()
```

```python
df=df.drop(['BsmtQual','FireplaceQu','GarageType'],axis=1)#here we drop the columns as we have created new feature by imputing
```
